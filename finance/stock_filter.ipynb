```python
import pandas as pd
import tushare as ts
```

```python
HS300 = ts.get_hs300s()
perfo = ts.get_report_data(2018,1)
```

合并

```python
result = pd.merge(HS300, perfo, how='inner', on=['code'])
```

排序，去重

```python
data = result.sort_values(by='eps', ascending=False)
data.drop_duplicates('code')
```
