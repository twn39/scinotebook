```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (14, 6)
import pandas as pd
plt.style.use('ggplot')
```

### 函数

$$y=ax + b$$

```python
x = np.linspace(-5, 5, 200)
y = lambda x : 3 * x -5
plt.plot(x, y(x), color='green', label='$y=3x-5$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=3x-5$'])
```

$$y=ax^2 + bx +c$$
它的对称轴为直线
$$x=-\frac{b}{2a}$$

```python
x = np.linspace(-5, 7, 200)
y = lambda x : 2 * (x**2) - 3 * x -5

m = (3/4) * np.ones(20)
g = np.linspace(-20, 80, 20)

plt.plot(x, y(x), m, g)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=2x^2-3x-5$', '$x=3/4$'])
```

$$y=\frac{a}{x}$$

```python
x = np.linspace(-2, 2, 200)
y = lambda x : 3/x
plt.plot(x, y(x), color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=3/x$'])
```

$$y=asin(x)$$

```python
x = np.linspace(-10, 10, 200)
y = lambda x : 2 * np.sin(x)
plt.plot(x, y(x), color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=2sin(x)$'])
```

$$y=acos(x)$$

```python
x = np.linspace(-10, 10, 200)
y = lambda x : 2 * np.cos(x)
plt.plot(x, y(x), color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=2cos(x)$'])
```

$$y=atan(x)$$

```python
x = np.linspace(-3, 3, 200)
y = lambda x : 2 * np.tan(x)
plt.plot(x, y(x), color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=2tan(x)$'])
```

$$y=a^x$$

```python
x = np.linspace(-5, 5, 200)
y = lambda x : 3**x
plt.plot(x, y(x), color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=3^x$'])
```

$$y=log_ax$$

```python
x = np.linspace(0.1, 3, 200)
y = lambda x : np.log2(x)
plt.plot(x, y(x), color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=log_2x$'])
```
