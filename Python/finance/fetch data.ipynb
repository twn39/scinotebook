```python
%pylab inline
matplotlib.style.use('ggplot')
import tushare as ts
import pandas as pd
```

通过tushare获取股票数据

```python
data = ts.get_k_data('002466', start='2017-07-01', end='2018-07-20')
data
```

处理数据,修改索引

```python
data = data.set_index('date')
data.index = pd.to_datetime(data.index)
data
```

```python
data['close'].plot()
ylabel('price')
title('close price of 002466.sz')
legend(['close price'])
```

数据合并

```python
hs300 = ts.get_hs300s()
all_report_data = ts.get_report_data(2018,1)
```

```python
result = pd.merge(hs300, all_report_data, how='inner', on='code')
result
```

根据eps倒叙排序

```python
result.sort_values(by='eps',  ascending=False)
```
