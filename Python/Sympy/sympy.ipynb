```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (14, 6)
plt.style.use('ggplot')
from sympy import *
## 美化输出
init_printing(use_unicode=True)
```

解方程:  
$$y=x^3 + x -2$$

```python
x = symbols('x')
exp = x**3 + x -2
solve(exp, x)
```

画图

```python
from sympy.plotting import plot
plot(exp, (x, -5, 5))
```

```python
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')

p1 = plot(x*x, (x, -10, 10), line_color="grey", show=False)
p2 = plot(x,(x, -10, 10), show=False)

p1.append(p2[0])
p1.show()
```

```python
p2 = plot( 
    ((x * x), (x, -10, 10)),
    (x, (x, -10, 10)),
    xlim=(-5, 10),
    ylim=(-5, 10),
)
```

```python
import numpy as np
plt.figure(figsize=(14, 6), layout='constrained')
a = np.linspace(-5, 10, 100)
b = np.linspace(0, 10, 150)
plt.plot(a, a, '--', label='$y=x$', color="grey") 
plt.plot(b, b**2, label='$y=x^2$')  
plt.plot(b, np.sqrt(b), label='$\sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title("Simple Plot")
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.legend()
plt.show()
```
