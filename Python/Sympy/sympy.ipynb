```python
%pylab inline
import matplotlib as plt
plt.rcParams["figure.figsize"] = (14, 6)
plt.style.use('ggplot')
from sympy import *
## 美化输出
init_printing(use_unicode=True)
```

解方程  $y=x^2-2$

```python
x = symbols('x')
exp = x**3 + x -2
solve(exp, x)
```

画图

```python
from sympy.plotting import plot
plot(exp, (x, -5, 5))
```
