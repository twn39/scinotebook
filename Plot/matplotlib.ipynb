```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (14, 6)
plt.style.use('ggplot')
import numpy as np
```

```python
x = np.linspace(-5, 5, 150)
plt.figure(figsize=(14, 6), layout='constrained')
plt.plot(x, x**3 + 3*x - 4, '-.', color='b', label="$x^3+3x-4$")
plt.plot(x, x**2 + 5*x -10, color='r', label="$x^2+5x-3$")
plt.legend() ## 加入注释
plt.xlabel('x')
plt.ylabel('y')
plt.title('Matplotlib Plot')
plt.xlim(-2, 5)
plt.ylim(-50, 50)
plt.show()
```
