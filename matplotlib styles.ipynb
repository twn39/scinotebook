```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# 获取所有样式
styles = plt.style.available

# 为每个样式生成图表
for style in styles:
    plt.style.use(style)  # 应用样式
    plt.figure()  # 创建新图表
    plt.plot(x, y, label='Line 1')
    plt.title(f'Style: {style}', fontsize=14)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()
```
